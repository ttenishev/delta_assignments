## Approach 2: Random Forest

Random Forest is capable to make predictions using information from features, but it never accounts for possible autocorrelation between observations. We can still try to include Y values of past observations as Xs of current observations. Let's see how well it goes

````python
from sklearn.ensemble import RandomForestRegressor
````

````python
train_df_w_lags = train_df.copy().drop('Date', axis = 1)

n_lags = 2

for i in range(1, n_lags + 1):
    train_df_w_lags[f'{target_var_name}-{i}'] = train_df_w_lags[target_var_name].shift(i)
    
train_df_w_lags.dropna(inplace = True)
````

````python
X = train_df_w_lags.drop([target_var_name], axis = 1)
y = train_df_w_lags[target_var_name]
````

````python
RF = RandomForestRegressor(n_estimators=100, random_state=1)
fit = RF.fit(X, y)
y_pred = fit.predict(X)
````

````python
RF_mse = mean_squared_error(y, y_pred)
print('RF RMSE:', RF_mse**0.5)
````

RF RMSE without train_test splitting is three times smaller than ARIMA RMSE with train-test split. For now this is an improvement, let's see how it goes with the split

````python
train_size = int(len(y) * 0.8)
X_train, X_test = X[0:train_size], X[train_size:len(y)]
y_train, y_test = y[0:train_size], y[train_size:len(y)]

RF_w_split = RandomForestRegressor(n_estimators=100, random_state=1)
fit = RF_w_split.fit(X_train, y_train)
y_pred_w_split = fit.predict(X_test)

RF_w_split_mse = mean_squared_error(y_test, y_pred_w_split)
print('RF RMSE:', RF_w_split_mse**0.5)
````

Test RMSE of RF is twice as large as of train but still better than of ARIMA

````python
# Single step dataset preparation
def singleStepSampler(df, window):
	xRes = []
	yRes = []
	for i in range(0, len(df) - window):
		res = []
		for j in range(0, window):
			r = []
			for col in df.columns:
				r.append(df[col][i + j])
			res.append(r)
		xRes.append(res)
		yRes.append(df[[target_var_name]].iloc[i + window].values)
	return np.array(xRes), np.array(yRes)
````

## <font color = 'green'>Approach 5: Feed Forward Neural Network (Best model) </font>

````python
from sklearn.preprocessing import StandardScaler

target_var_name = train_df.columns[-1]

x_train = train_df.drop(['Date', target_var_name], axis = 1)
y_train = train_df[target_var_name]

scaler = StandardScaler()
scaler = scaler.fit(x_train)
df_for_training_scaled = scaler.transform(x_train)
````

````python
from tensorflow import keras
from tensorflow.keras import layers
#Build the model:

m = x_train.shape[1] #number of features

model = keras.Sequential([
    layers.Dense(450, activation="relu", kernel_initializer = 'he_normal', input_shape = (m,)),
    layers.Dropout(0.1),
    layers.Dense(450, activation="relu", kernel_initializer = 'he_normal'),
    layers.Dropout(0.1),
    layers.Dense(450, activation="relu", kernel_initializer = 'he_normal'),
    layers.Dropout(0.1),
    layers.Dense(450, activation="relu", kernel_initializer = 'he_normal'),
    layers.Dropout(0.1),
    layers.Dense(450, activation="relu", kernel_initializer = 'he_normal'),
    layers.Dropout(0.1),
    layers.Dense(1) # it does not have activation (linear layer by default)
])

model.summary()
````

````python
#compile the model:
model.compile(
    optimizer=keras.optimizers.Adam(0.001),
    loss=keras.losses.MeanSquaredError(),
    metrics=[
        keras.metrics.MeanAbsoluteError(),
    ],
)
````

````python
#Split into train and validation set:
from sklearn.model_selection import train_test_split

x_train_, x_val, y_train_, y_val = train_test_split(df_for_training_scaled, y_train, test_size = 0.2, random_state = 66)
````

````python
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10,mode='min',min_delta=0.5)

# Fit the model:
history = model.fit(x=x_train_,
    y=y_train_,
    batch_size=16,
    epochs=200,
    callbacks = [early_stopping],
    validation_data = (x_val, y_val))
````

````python
#Plot the training and the validation loss using the training history from the fitted model:
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label = 'Training Loss')
plt.plot(history.history['val_loss'], label = 'Validation Loss')
plt.legend()
````

````python
results_train = model.evaluate(x_train_, y_train_, verbose=0)
results_val = model.evaluate(x_val, y_val, verbose=0)

print('MSE train set: ',results_train[0])
print('MSE val set:',results_val[0],'\n')

print('MAE train set: ',results_train[1])
print('MAE val set: ',results_val[1])
````

Reorganize the data into a suitable format for model training. Essentially, here we create sequences that are going to be the input data for our RNN

````python
def create_dataset(time_series, timesteps):
    """
    Creates sequences of data for training/testing an LSTM model.

    Parameters:
    time_series (numpy array): The input time series data where each row is a time step and columns are features.
    timesteps (int): The number of timesteps to include in each sequence.

    Returns:
    numpy array: Array of sequences (X) with shape (num_sequences, timesteps, num_features).
    numpy array: Array of target values (Y) corresponding to each sequence.

    Description:
    This function takes a time series dataset and creates sequences of a specified number of timesteps. 
    Each sequence (X) includes 'timesteps' number of rows from the input time series. The corresponding 
    target value (Y) for each sequence is the value of the target variable at the timestep immediately 
    following the end of the sequence.

    Example:
    If the time_series array has the shape (100, 5) and timesteps is set to 7, the output will be:
    - dataX: Array with shape (93, 7, 5), where 93 is the number of sequences, 7 is the number of timesteps, 
      and 5 is the number of features.
    - dataY: Array with shape (93,), containing the target values for each sequence.
    """
    dataX, dataY = [], []
    
    for i in range(0, len(time_series) - timesteps):  # Iterate over the time series data
        x = time_series[i:(i + timesteps)]  # Extract a sequence of length 'timesteps'
        dataX.append(x)  # Append the sequence to dataX
        y = time_series[(i + timesteps)][-1]  # Extract the target value at the next timestep
        dataY.append(y)  # Append the target value to dataY
           
    return np.array(dataX), np.array(dataY)  # Convert lists to numpy arrays and return
````

The **timesteps** parameter defines how many steps back in time should be used to predict the target value. It is quite similar the number of lags ```q``` in the AR(q). Actually, we even pick the timesteps just as we define the ```q``` in usual AR processes, based on the partial autocorrelation function.

## Step 4: Training the model

As an evaluation criterion, **RMSE** is used. It penalizes large errors and the scores are easier to interpret as the units are of the same scale as our target variable.

As for **benchmark** models, we will use:
1. a naive approach called the persistence forecast, i.e. next step is predicted according to the value of the previous step;
2. the best model from assignment 1 - there it was just a feedforward neural network that did not account for time series component in the data.

````python
from sklearn.metrics import mean_squared_error
def rmse(y_true, y_score):
    error = np.sqrt(mean_squared_error(y_true=y_true, y_pred=y_score))
    return error
````

````python
rmse_train = {}
rmse_val = {}
````

### Step 4.1: Persistence forecast

To make the naive benchmark as good as possible, we will optimize the lag, so that $\hat{x}_{t} = x_{t-k}$ for the optimal $k$.

````python
persistence_values = range(1, 23) # how many days back (within a month) should we go as a forecast for tomorrow
scores = list()
for p in persistence_values:
    # walk-forward validation
    history = [x for x in train[:,-1]]
    predictions = list()
    for i in range(len(val)):
    # make prediction
        yhat = history[-p]
        predictions.append(yhat)
    # observation
        history.append(val[i][-1])
    # report performance
    error = rmse(val[:,-1], predictions)
    scores.append(error)
    print('p=%d RMSE:%.5f' % (p, error))
# plot scores over persistence values
plt.plot(persistence_values, scores)
plt.show()
````

Judging from the plot, the lag $k=1$ is the best for the naive approach\
Since the persistence forecasts look at a past value of the time series, it is convenient to use the Pandas function `shift()` for calculating the benchmark. So we first collect the actual time series in a data frame, and then use shift with our selected lag k_naive, while filling up NaN values at the beginning of the time series with the mean. Since we will never use the start of the time series (i.e., first samples in the training set) for testing, the mean replacement does not affect subsequent results.

````python
# Lag for the naive benchmark (i.e., persistence forecast)
k_naive = 1
pred_naive = pd.DataFrame(np.vstack([y_train, y_val])) 
pred_naive = pred_naive.shift(periods=k_naive, axis=0, fill_value=np.mean(y_train)).values
````

It is good practice to collect relevant variables and results in one data structure. This ensures that we have all the data we need in one place. We will create a dictionary in which we store the true realizations of our target variable (i.e., daily price differences) and the predictions of various models. This will make it easy to compare different models later on.

Further, to easy the interpretations of error values, it is preferable to reverse the scaling of the time series. For example, this allows us to interpret a root mean squared-error value as a deviation between the actual price difference and our forecast. Should we consider trading on our model - and you should not consider trading on any mode we build today, we would want to examine actual deviations as opposed to deviations between scaled variables. Reversing a scaling is very easy with scikit-learn.

````python
pred_naive[:len(y_train), :].shape
````

Since we used one scaler for the entire dataset, we will need to pass the input of the same dimension to the inverse scaler. For that, we just stack a matrix of zeroes and our target, pass them to the inverse scaler and then take the target from the unscaled matrix

````python
# We use one dictionary for the training and one for the test data
pred_val = {}
pred_train = {}

zeros_train = np.zeros((y_train.shape[0], train.shape[1]-1))
zeros_val = np.zeros((y_val.shape[0], val.shape[1]-1))

# We store the actual price differences as opposed to their scaled version
pred_train["true"] = scaler.inverse_transform(np.hstack([zeros_train, y_train]))[:,-1]  # store as 1d array
pred_val["true"]  = scaler.inverse_transform(np.hstack([zeros_val, y_val]))[:,-1]

# Also add the lagged persistence forecast
pred_train["naive"] = scaler.inverse_transform(np.hstack([zeros_train, pred_naive[:len(y_train), :]]))[:,-1]  # This is a good example of why the [a, b-1] 
pred_val["naive"]  = scaler.inverse_transform(np.hstack([zeros_val, pred_naive[len(y_train):, :]]))[:,-1]  # indexing is convenient
````

Let's plot predictions and actuals for illustration. To that end, it is nice that we can easily convert our dictionary into a data frame and use the corresponding plotting routines.

````python
pd.DataFrame(pred_train).plot()
plt.show()
````

Calculate the RMSE of the naive model as a first indication.

````python
rmse_train["naive"] = rmse(pred_train["true"], pred_train["naive"])
rmse_val["naive"]  = rmse(pred_val["true"],  pred_val["naive"])

rmse_val
````

### Step 4.3: LSTM with multiple layers

Let's try to add more layers and increase the number of nodes, as well as ad dropout rates as regularization to avoid overfitting.
Note: the learning rate was also changed to a lower one, as the graphs from the next steps clearly showed the gradient updates were overjumping the (local?) minimum when the model parameters were tuned

````python
stateless_multilayer = Sequential()
stateless_multilayer.add(LSTM(100, return_sequences=True))
stateless_multilayer.add(Dropout(0.2))  # you can also add dropout layers if you like
#stateless_multilayer.add(LSTM(100, return_sequences=True))
#stateless_multilayer.add(Dropout(0.2))
#stateless_multilayer.add(LSTM(100, return_sequences=True))
#stateless_multilayer.add(Dropout(0.2))
stateless_multilayer.add(LSTM(100, return_sequences=False))
stateless_multilayer.add(Dropout(0.2))
stateless_multilayer.add(Dense(1))
stateless_multilayer.compile(loss = 'mean_squared_error', optimizer = keras.optimizers.Adam(0.000002), metrics=['mae'])

#early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10,mode='min',min_delta=0.5)
````

The first layer uses *return_sequences=True*. Hence, we get one output value for each time step. These outputs can then be used as input for the second LSTM layer. Same goes for other layers, except for the top one. The output of one layer is the input of the next layer.

The top layer has *return_sequences=False*. This means that we only get one forecast for the target variable from the last hidden state after processing the full sequence. Therefore, we do not need to adjust our target in any way and can train right with the 2d array *y_train* as follows. 

````python
history = stateless_multilayer.fit(X_train, y_train, batch_size=batch_size*8, epochs=50,
                                   validation_data = (X_val, y_val), shuffle = False)
````
